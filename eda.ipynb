```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
def eda(nombre_archivo):
    nombre_archivo = pd.read_csv(f'{nombre_archivo}.csv')

    print(f"\nAN√ÅLISIS EXPLORATORIO DE DATOS")
    print(f"\n{nombre_archivo.shape[0]} FILAS Y {nombre_archivo.shape[1]} COLUMNAS")
    if nombre_archivo.duplicated().any():
        print("‚ö†Ô∏èDUPLICADOS:", nombre_archivo.duplicated().sum())
    print("\nCOLUMNAS:\n", nombre_archivo.columns, "\n")
   
    contador = 0
    for i in nombre_archivo.columns:
        print("\n-----------------------------------------------------------------")
        print("üü¶ COLUMNA", contador, ": ", i)
        nulos = nombre_archivo[i].isnull()
        if nulos.any():
            print("\n\n‚ö†Ô∏èNULOS:")
            print(nulos.sum())
        else:
            print("\nNO HAY VALORES NULOS")
        if nombre_archivo[i].value_counts().count() < 30:
            print(nombre_archivo[i].value_counts().count(), "VALORES:")
            print(nombre_archivo[i].value_counts().sort_values())
        else:
            print("\n‚ö™NUMEROSOS VALORES, POR EJEMPLO:")
            print(nombre_archivo[i].sample(5))
        contador +=1
    print("--------------------------------------------------------------------")

    display(nombre_archivo.head())
    display(nombre_archivo.tail())
    display(nombre_archivo.sample(5))


    print("\nINFORMACI√ñN GENERAL:")
    print(nombre_archivo.info())

    print("\nTIPOS DE DATOS:")
    print(nombre_archivo.dtypes)

    print("\nDESCRIPCI√ìN ESTAD√çSTICA SIN LA COLUMNA DE ID 'Loyalty Number' (ya que no es verdaderamente num√©rica):")
    descripcion = nombre_archivo.describe(include='all').round(2).T
    descripcion_sin_falsas_numericas = descripcion.drop(['Loyalty Number', 'Year', 'Month'], errors='ignore')
    display(descripcion_sin_falsas_numericas)
```

```python
eda('Customer Flight Activity')
```

```python
eda('Customer Loyalty History')
```
